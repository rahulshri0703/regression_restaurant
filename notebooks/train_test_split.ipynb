```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['rate']),df['rate'],test_size=0.2,
                                                 random_state=66,stratify=df['listed_in(type)'])

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train_s = pipe.transform(x_train)
x_test_s = pipe.transform(x_test)
x_val_s = pipe.transform(x_val)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
lr = LinearRegression()
lr.fit(x_train_s,y_train)
```

```python
y_train_pred = lr.predict(x_train_s)
y_test_pred = lr.predict(x_test_s)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```
